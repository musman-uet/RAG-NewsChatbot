```python
import os
import time
import threading

# Name of the streamlit app file
streamlit_file = "projectphase1_streamlitapp2.py"

# Start Streamlit app
def run_streamlit():
    os.system(f"streamlit run {streamlit_file}")

# Start LocalTunnel
def run_localtunnel():
    os.system("lt --port 8501 > tunnel.txt &")

# Launch both
threading.Thread(target=run_streamlit).start()
time.sleep(5)

threading.Thread(target=run_localtunnel).start()
time.sleep(5)

# Print public URL
!tail -n 1 tunnel.txt
```

```python
import os
import time
import threading
from pyngrok import ngrok

# Make sure old tunnels are closed
ngrok.kill()

# Set Streamlit port
port = 8502

# Start ngrok tunnel
public_url = ngrok.connect(port)
print(f" Streamlit is running at: {public_url}")

# Function to run streamlit
def run():
    os.system(f"streamlit run projectphase1_streamlitapp2.py --server.port {port} --server.enableCORS false")

# Start Streamlit in a new thread
thread = threading.Thread(target=run)
thread.start()

# Wait for app to fully load
time.sleep(10)
```
